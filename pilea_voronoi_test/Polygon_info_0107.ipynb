```python
info_array = np.zeros((25, 4))
```

```python
def area_sum_n_num_edge(faces):
    area_sum = 0
    faces = G.graph['faces_passed']
    num_edge_list = []
    for face in faces:
        num_edge_list.append(len(face))
        face_poly = Polygon(face)
        area_sum += face_poly.area
    ave_edge_num = np.array(num_edge_list).mean()
    
    return area_sum, ave_edge_num
```

```python
for i, sample in enumerate(sample_list):
    G = graph_create.graph_creation(sample)
    info_array[i][0] = len(G.graph['faces'])
    info_array[i][1] = len(G.graph['faces_passed'])
    area_sum, ave_edge_num = area_sum_n_num_edge(G.graph['faces_passed'])
    blade_area = Polygon(G.graph['boundary']).area
    info_array[i][2] = area_sum/blade_area
    info_array[i][3] = ave_edge_num
```
