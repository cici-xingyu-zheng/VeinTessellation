```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

The idea of this exploration is to think about the influence of auxin sources like some gravitational force exerted by multiple astronomical objects.

There are two ideas to plot one is to plot the vector field directly, and another is to plot some concentration of auxin in 2D, and use some derivative of that to reprensent directional pulls, like a gradient vector field.

```python
fig, ax = plt.subplots(figsize = (12,12))

x,y = np.meshgrid(np.linspace(-5,5,20),np.linspace(-5,5,20))


u = - (x+2)/np.sqrt((x+2)**2 + (y-3)**2)**3 -  (x-4)/np.sqrt((x-4)**2 + y**2)**3 -  (x+2)/np.sqrt((x+2)**2 +  (y+3)**2)**3
v = - (y-3)/np.sqrt((x+2)**2 + (y-3)**2)**3 -  y/np.sqrt((x-4)**2 + y**2)**3 -  (y+3)/np.sqrt((x+2)**2 +  (y+3)**2)**3




length = np.sqrt(u**2 + v**2)

u = np.where(length > 1, 0, u )
v = np.where(length > 1, 0, v )
# u = - (x-3)/np.sqrt((x-3)**2 + (y-3)**2) -  (x+3)/np.sqrt((x+3)**2 + y**2) 
# v = - (y-3)/np.sqrt((x-3)**2 + (y-3)**2) -  y/np.sqrt((x+3)**2 + y**2) 

color_array = np.sqrt(v**2 + v**2)

p = ax.quiver(x,y,u,v,color_array, alpha=0.8)
plt.show()
```
